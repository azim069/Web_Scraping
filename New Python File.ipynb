```python
data = []
url = 'https://www2.hm.com/en_us/women/new-arrivals/clothes/_jcr_content/main/productlisting.display.json?sort=stock&image-size=small&image=model&offset=0&page-size=700'
headers={
    'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36 OPR/94.0.0.0'
    }
page = requests.get(url,headers=headers)
response=page.json()
for i in range(700):
    products={
    "title" : response['products'][0]['title'],
    "price" : '{:.2f}'.format(float(response['products'][i]['price'].split(' ')[1])*107),
    "image" : response['products'][0]['image'] ,
    "sellingAttribute" : response['products'][0]['sellingAttribute'],
    "brandName" : response['products'][0]['brandName'],
    "link" : "https://www2.hm.com"+ response['products'][0]['link']   
    }
    data.append(products)
```

```python
DF = pd.DataFrame(data)
```
